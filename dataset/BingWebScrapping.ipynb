```python

from selenium import webdriver
from bs4 import BeautifulSoup
import requests

from pathlib import *
import time

import urllib.request
import os
myDirectory = "C:/Users/Charlne/Documents/Jedha/Projet" 
projet_path = Path(myDirectory)

gecko_path = projet_path/'geckodriver.exe'

recherche = 'homme visage'

driver = webdriver.Firefox(executable_path=gecko_path.as_posix())
recherche = 'Paris'
url= "http://www.bing.com/images/search?q=" + recherche + "&FORM=HDRSC2"
driver.get(url)


# Grce  cette boucle, on demande  Selenium d'effectuer 5 fois les actions suivantes:
    # Scroll la page des rsultats
    # Pause d'une demi-seconde

for i in range(100):
    driver.execute_script("scrollBy(0,1000);")
    time.sleep(2)

# Puis : Selenium cherche le tag "img" dans les rsultats d'images et liste tous les tags dtcts dans la page
time.sleep(10)
tag_img = driver.find_elements_by_tag_name('img')


# Dans le dossier du notebook, on cr un dossier "Paris_scrap"

try:
    os.mkdir('Visages_hommes')
except FileExistsError:
    pass

# On initialise un compteur  0 afin que chaque itration de la boucle ajoute 1 au compteur et permette de numroter les images rcupres
count = 0
err=0
# On itre sur l'ensemble des images contenues dans la variable "tag_img"

for i in tag_img:
    # On rcupre le contenu de l'attribut "src" de chacune des images
    src = i.get_attribute('src')
    
    # On rcupre le contenu de l'attribut "alt" de chacune des images
    alt = i.get_attribute('alt')
    
    # Si vrifie que l'attribut "src" est non nul et que l'attribut "alt" correspond  'Rsultat dimages pour + objet recherch 
    if src != None and alt == 'Rsultat dimages pour {}'.format(recherche):
        # On transforme l'URL rcupre dans "src" en une chane de caractre
        src  = str(src)
        print(src)
        # On ajoute 1 au compteur
        count+=1
        # Grce  la mthode "urlretrieve" d'Urllib, on enrengistre sous format .jpg le contenu de l'URL contenue dans "src"
        urllib.request.urlretrieve(src, os.path.join('Visages_hommes','Homme_'+str(count)+'.jpg'))
    
    else:
        err +=1
        continue

driver.quit()
```
